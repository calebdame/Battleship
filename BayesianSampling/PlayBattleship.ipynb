# Play Statistically Augmented Battleship!

### First, we have the Autoplay Object

With this BattleshipAutoplay object, you just have to give it a **loadingtime** during which it samples random boards and it will play each round, each time selecting the next  grid location based only off of the calculated probability.  You'll find that the **loadingtime** does not need to be very long to get good results, and a large increase may only bring marginal improvements in the number of turns it takes to solve

```python
from Battleship import Battleship, BattleshipPlayer, BattleshipAutoplay
from matplotlib import pyplot as plt
import numpy as np

dimentions = 10
ships = [2,3,3,4,5]
loadingtime = .05 #seconds
bins = 25
n = 100

turns = []
game = BattleshipAutoplay(dim=dimentions, ships=ships, lag=loadingtime)
for i in range(n):
    turns += [game.play(refresh=True)]
    
print("Mean:", np.mean(turns))
print("Variance:", np.var(turns))

plt.hist(turns, bins=bins)
plt.show()
```

# BattleshipPlayer

An visual and interactive way to see what is going on behind the scenes of the **BattleshipAutoplay** class. The only difference is the ability to visualize the board (and the heatmap of estimated probablities) and play yourself!

```python
battleshipInteractive = BattleshipPlayer(dim=dimentions, ships=ships, lag=1)
```

```python
battleshipInteractive.view(showGraph=True)
```

```python
battleshipInteractive.guess((5,4))
battleshipInteractive.view(showGraph=True)
```
